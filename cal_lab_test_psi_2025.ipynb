# Calendario degli incontri

Il calendario didattico comprende 14 esercitazioni della durata di 3 ore ciascuna e, negli ultimi due incontri, una verifica tramite Quiz Moodle e le presentazioni finali degli studenti.

## Laboratorio 1
- **Data:** 4 marzo 2025
- **Argomento:** Introduzione al corso e a `R`.
- **Orario:** 10:30-13:30

---

## Laboratorio 2
- **Data:** 6 marzo 2025
- **Argomento:** Esplorazione dei dati (EDA).
  - A. Discussione dell'articolo di @dima2018scale.
  - B. Importazione dei dati in `R` e analisi delle distribuzioni di densità.
  - C. Statistiche descrittive, curtosi e asimmetria.
  - D. Identificazione di outlier e calcolo della distanza di Mahalanobis.
  - E. Analisi del "Careless Responding" con presentazione degli articoli di @ward2023dealing e @stosic2024careless.
  - F. Applicazione dei principi del "Careless Responding" sui dati FAI.
- **Orario:** 8:30-11:30

---

## Laboratorio 3
- **Data:** 11 marzo 2025
- **Argomento:** Teoria Classica dei Test.
  - A. Modello di regressione lineare bivariato (coeff. minimi quadrati, errore standard, scomposizione della varianza e R²).
  - B. Regressione multipla: interpretazione dei coefficienti parziali e parziali di correlazione.
  - C. Introduzione alla CTT con una simulazione.
  - D. Calcolo delle quantità della CTT per i dati FAI (attendibilità, SEM, stima del punteggio vero con il metodo di Kelley, IC).
  - E. Discussione dell’articolo di @blampied2022reliable.
  - F. Limiti della SEM nella CTT.
  - G. Presentazione degli articoli di @grund2024multiple e @van2020rebutting.
  - H. Applicazione dell’imputazione multipla ai dati FAI.
- **Orario:** 10:30-13:30

---

## Laboratorio 4

  - A. Introduzione alla regressione logistica.
  - B. Applicazione ai dati FAI.
  - C. AUC.
  - D. Introduzione alla Mokken Scale Analysis.
  - E. Applicazione della MSA ai dati FAI.
- **Orario:** 8:30-11:30

## Laboratorio 5

  - G. Applicazione degli STANDARDS al FAI.
